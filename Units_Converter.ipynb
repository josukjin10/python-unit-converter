```python
user_input = ''

while user_input != "exit":

  measurement = ["", "Distance/Length", "Temperature", "Mass", "Volume","Time"]

  print("----Units Converter----")
  print(f"1. {measurement[1]}")
  print(f"2. {measurement[2]}")
  print(f"3. {measurement[3]}")
  print(f"4. {measurement[4]}")
  print(f"5. {measurement[5]}")
  print()
  choice = (input("Which measurement would you like to convert?: "))
  print()
  valid_choices = ["1","2","3","4","5"]

  if choice not in valid_choices:   # If the user types a character other than the valid choices, this stops the program.
    print("Invalid selection! You must enter 1, 2, 3, 4, or 5.")
    continue

# divider =======================================================================================

  if choice == "1":                 # If the user selects "Distance"
      units = ["","Kilometers","Meters","Centimeters","Miles","Yards","Feet","Inches"]
      print("----Distance/Length----")
      print(f"1. {units[1]}")
      print(f"2. {units[2]}")
      print(f"3. {units[3]}")
      print(f"4. {units[4]}")
      print(f"5. {units[5]}")
      print(f"6. {units[6]}")
      print(f"7. {units[7]}")
      print()
      unit_from = (input("Which unit would you like to convert from?: "))
      print()

      valid_choices = ["1","2","3","4","5","6","7"]
      if unit_from not in valid_choices:  # If the user types a character other than the valid choices, this stops the program.
          print("Invalid selection! You must enter 1, 2, 3, 4, 5, 6, or 7")
          continue
      else:
          unit_1 = units[int(unit_from)]    # Changes the string "unit_from" into an integer and indexes the correct selection from the "units" list. "1" --> Kilometers
          print(f"1. {units[1]}")
          print(f"2. {units[2]}")
          print(f"3. {units[3]}")
          print(f"4. {units[4]}")
          print(f"5. {units[5]}")
          print(f"6. {units[6]}")
          print(f"7. {units[7]}")
          print()
          unit_to = (input("Which unit would you like to convert to?: "))
          print()

          valid_choices = ["1","2","3","4","5","6","7"]

          if unit_to not in valid_choices:
              print("Invalid selection! You must enter 1, 2, 3, 4, 5, 6 or 7.")
              continue
          elif unit_from == unit_to:
              print("Are you stupid?")
              continue
          else:
              unit_2 = units[int(unit_to)]
              number = float(input("Enter the number you want to convert: "))
              print()
              meters_in_unit = [0, 1000, 1, 0.01, 1609.34, 0.9144, 0.3048, 0.0254]

              rate_from = meters_in_unit[int(unit_from)]
              rate_to = meters_in_unit[int(unit_to)]

              amount_in_meters = number * rate_from
              final_conversion = amount_in_meters / rate_to

              print(f"Result: {round(final_conversion,4)} {unit_2}")

#=================================================================================================

  elif choice == "2":         #If the user selects "Temperature"
    units = ["","˚C","˚F","K"]
    print("----Temperature----")
    print(f'1. {units[1]}')
    print(f'2. {units[2]}')
    print(f'3.  {units[3]}')
    print()
    unit_from = (input("Which unit would you like to convert from?: "))
    print()

    valid_choices = ["1","2","3"]
    if unit_from not in valid_choices:
      print("Invalid selection! You must enter 1, 2, or 3")
      continue
    else:
      unit_1 = units[int(unit_from)]
      print(f'1. {units[1]}')
      print(f'2. {units[2]}')
      print(f'3.  {units[3]}')
      print()
      unit_to = input("Which unit would you like to convert to?: ")
      print()


      valid_choices = ["1","2","3"]
      if unit_to not in valid_choices:
        print("Invalid selection! You must enter 1, 2, or 3")
        continue
      else:
        number = float(input("What number would you like to convert?: "))
        unit_2 = units[int(unit_to)]

        if unit_from == unit_to:
          print("Are you stupid?")
          continue
        elif unit_from == "1" and unit_to == "2":     # Converting Celsius into Fahrenheit
          celsius = number
          final_conversion = (celsius*9/5)+32

          print(f'Result: {round(final_conversion,4)} {unit_2}')

        elif unit_from =='1' and unit_to == "3":
          celsius = number
          final_conversion = celsius + 273.15

          print(f'Result: {round(final_conversion,4)} {unit_2}')

        elif unit_from == '2' and unit_to == "1":
          celsius = (number-32)*5/9
          final_conversion = celsius

          print(f'Result: {round(final_conversion,4)} {unit_2}')

        elif unit_from =='2' and unit_to == '3':
          celsius = (number-32)*5/9
          final_conversion = celsius + 273.15

          print(f'Result: {round(final_conversion,4)} {unit_2}')

        elif unit_from == '3' and unit_to == '1':
          celsius = number - 273.15
          final_conversion = celsius

          print(f'Result: {round(final_conversion,4)} {unit_2}')

        elif unit_from == '3' and unit_to == '2':
          celsius = number - 273.15
          final_conversion = (celsius*9/5)+32

          print(f'Result: {round(final_conversion,4)} {unit_2}')

#=================================================================================================

  elif choice == "4":     #If the user selects "Volume"
      units = ["", "Cubic Meters", "Liters", "Milliliters", "Gallons","Quarts","Cubic Inches"]
      print("----Volume----")
      print(f'1. {units[1]}')
      print(f'2. {units[2]}')
      print(f'3. {units[3]}')
      print(f'4. {units[4]}')
      print(f'5. {units[5]}')
      print(f'6. {units[6]}')
      print()
      unit_from = (input("Which unit would you like to convert from?: "))

      valid_choices = ["1",'2','3','4','5','6']

      if unit_from not in valid_choices:
        print("Invalid selection! You must enter 1, 2, 3, 4, 5, or 6")
        continue
      else:
        unit_1 = units[int(unit_from)]
        print()
        print(f'1. {units[1]}')
        print(f'2. {units[2]}')
        print(f'3. {units[3]}')
        print(f'4. {units[4]}')
        print(f'5. {units[5]}')
        print(f'6. {units[6]}')
        print()
        unit_to = input("Which unit would you like to convert to?: ")
        print()


      if unit_to not in valid_choices:
        print("Invalid selection! You must enter 1, 2, 3, 4, 5, or 6")
        continue

      elif unit_from == unit_to:
        print("Are you stupid?")
        continue

      else:
        unit_2 = units[int(unit_to)]
        number = float(input("What number would you like to convert?: "))
        print()
        liters_in_unit = [0, 1000, 1, 0.001, 3.78541, 0.946353, 0.0163871]

        rate_from = liters_in_unit[int(unit_from)]
        rate_to = liters_in_unit[int(unit_to)]

        amount_in_liters = number * rate_from
        final_conversion = amount_in_liters / rate_to

        print(f'Result: {round(final_conversion,4)} {unit_2}')

#=================================================================================================


  elif choice == "3":         #If the user selects "Mass"
    units = ['', 'Kilograms','Grams','Milligrams','Pounds','Ounces']
    print("----Mass----")
    print(f'1. {units[1]}')
    print(f'2. {units[2]}')
    print(f'3. {units[3]}')
    print(f'4. {units[4]}')
    print(f'5. {units[5]}')
    print()
    unit_from = (input("Which unit would you like to convert from?: "))
    valid_choices = ['1','2','3','4','5']

    if unit_from not in valid_choices:
      print("Invalid selection! You must enter 1, 2, 3, 4, or 5")
      continue
    else:
      unit_1 = units[int(unit_from)]
      print()
      print(f'1. {units[1]}')
      print(f'2. {units[2]}')
      print(f'3. {units[3]}')
      print(f'4. {units[4]}')
      print(f'5. {units[5]}')
      print()
      unit_to = input("Which unit would you like to convert to?: ")
      print()

      if unit_to not in valid_choices:
        print("Invalid selection! You must enter 1, 2, 3, 4, or 5")
        continue

      elif unit_from == unit_to:
        print("Are you stupid?")
        continue

      else:
        unit_2 = units[int(unit_to)]
        number = float(input("What number would you like to convert?: "))
        print()
        grams_in_unit = [0, 1000, 1, 0.001, 453.592, 28.3495]

        rate_from = grams_in_unit[int(unit_from)]
        rate_to = grams_in_unit[int(unit_to)]

        amount_in_grams = number * rate_from
        final_conversion = amount_in_grams / rate_to

        print(f'Result: {round(final_conversion,4)} {unit_2}')

#=================================================================================================

  elif choice == "5":
    units = ['','Weeks','Days','Hours','Minutes','Seconds','Milliseconds']
    print('----Time----')
    print(f'1. {units[1]}')
    print(f'2. {units[2]}')
    print(f'3. {units[3]}')
    print(f'4. {units[4]}')
    print(f'5. {units[5]}')
    print(f'6. {units[6]}')
    print()
    unit_from = (input("Which unit would you like to convert from?: "))

    valid_choices = ["1",'2','3','4','5','6']

    if unit_from not in valid_choices:
      print("Invalid selection! You must enter 1, 2, 3, 4, 5, or 6")
      continue
    else:
      unit_1 = units[int(unit_from)]
      print()
      print(f'1. {units[1]}')
      print(f'2. {units[2]}')
      print(f'3. {units[3]}')
      print(f'4. {units[4]}')
      print(f'5. {units[5]}')
      print(f'6. {units[6]}')
      print()
      unit_to = input("Which unit would you like to convert to?: ")
      print()

      if unit_to not in valid_choices:
        print("Invalid selection! You must enter 1, 2, 3, 4, 5, or 6")
        continue

      elif unit_from == unit_to:
        print("Are you stupid?")
        continue

      else:
        unit_2 = units[int(unit_to)]
        number = float(input("What number would you like to convert?: "))
        print()
        seconds_in_unit = [0, 604800, 86400, 3600, 60, 1, 0.001]

        rate_from = seconds_in_unit[int(unit_from)]
        rate_to = seconds_in_unit[int(unit_to)]

        amount_in_seconds = number * rate_from
        final_conversion = amount_in_seconds / rate_to

        print(f'Result: {round(final_conversion,4)} {unit_2}')

  print()
  user_input = input("If you wish to quit, type \"exit,\" and if you wish to continue, type anything: ")
  print()
```
